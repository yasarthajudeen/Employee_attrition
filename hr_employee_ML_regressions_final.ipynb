```python
df1=df.copy()
```

```python
df1.drop(['EmployeeCount','Over18','StandardHours','EmployeeNumber'],axis=1,inplace=True)
```

```python
df1.info()
```

```python
df1.head()
```

```python
df1_dummy=pd.get_dummies(df1,columns=['Attrition','BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus','OverTime'],drop_first=True)
```

```python
df1_dummy
```

```python
df2=df1_dummy.copy()
```

```python
df2.info()
```

```python
X=df2.drop('Attrition_Yes',axis=1)
y=df2['Attrition_Yes']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
lr_model=LinearRegression()
log_model=LogisticRegression()
svc_model=SVC()
scaler=StandardScaler()
knn_classifier=KNeighborsClassifier()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
lr_model.fit(X_train_scaled,y_train)
```

```python
lr_model.score(X_test_scaled,y_test)
```

```python
log_model.fit(X_train_scaled,y_train)
```

```python
log_model.score(X_test_scaled,y_test)
```

```python
svc_model.fit(X_train_scaled,y_train)
```

```python
svc_model.score(X_test_scaled,y_test)
```

```python
knn_classifier.fit(X_train_scaled,y_train)
```

```python
knn_classifier.score(X_test_scaled,y_test)
```
